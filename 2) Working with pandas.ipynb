## Accessing another table 

```python
app_user_reviews=pd.read_csv('D:\python notes\Python\data cleaning and visualization\googleplaystore_user_reviews.csv')
```

```python
app_user_reviews.head(2)
```

```python
#Creating that coloumn which will be the part of both tables and having a unique records in itself 
app_user_reviews['S NO']=range(1,len(app_user_reviews.index)+1)
```

```python
#Creating that coloumn which we want to merged in our existing table just for the pratice although this coloumn doesn't make sense
app_user_reviews['Valid_reviews']=''
```

```python
for i in app_user_reviews:
    app_user_reviews['Valid_reviews']=np.random.randint(0,100,size=(len(app_user_reviews.index)))
```

```python
app_user_reviews.head()# well randint is assigning same numbers to all the records thats bad  
```

Merging app_user_reviews table's table's coloumn 'Valid_reviews' in our app coloumn 

```python
result=pd.merge(app,
    app_user_reviews[['S NO','Valid_reviews']],
    on='S NO')
result.iloc[1:3,11:15]
```
